**Import Required Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

**Load the Iris Dataset**

```python
# Load the Iris dataset directly from seaborn
iris = sns.load_dataset('iris')

# Display first 5 rows
iris.head()
```

**Understand the Dataset Structure**

```python
# Check the shape (rows, columns)
print("Shape of dataset:", iris.shape)

# List column names
print("Column names:", iris.columns.tolist())

# Basic info about data types and nulls
print("\nDataset info:")
iris.info()

# Statistical summary
print("\nStatistical summary:")
iris.describe()
```

**Scatter Plot**

```python
# Scatter plot to see relationships between sepal length and width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

**Histograms**

```python
# Histogram for each numeric column
iris.hist(bins=20, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histogram of Iris Dataset Features', fontsize=16)
plt.tight_layout()
plt.show()
```

**Box Plots**

```python
# Box plots for each feature grouped by species
plt.figure(figsize=(12, 10))

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal Length by Species')

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title('Sepal Width by Species')

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal Length by Species')

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(data=iris, x='species', y='petal_width')
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

**Conclusion**

## Conclusion

- The Iris dataset contains 150 entries with 4 numeric features and 1 categorical target (`species`).
- Scatter plots reveal that **petal length and width** offer clear separation between species.
- Histograms show that features like `petal_length` and `petal_width` are skewed based on species.
- Box plots identified **Setosa** as having distinctly smaller petals compared to Versicolor and Virginica.
- This exploration builds foundational understanding of data loading, inspection, and visualization using `pandas`, `seaborn`, and `matplotlib`.
